<h1 style="color:green">Bag of Words Model</h1>

```python
import pandas as pd 
from sklearn.feature_extraction.text import CountVectorizer
```

- #### CountVectorizer transforms a collection of text documents into a matrix of token counts 
- #### it breaks the text into words(tokens) and counts how often each word appears
- #### this gives us the numerical representation we need to apply ML techniques to text data 

```python
data = [' Most shark attacks occur about 10 feet from the beach since that is where the people are',
        'the efficiency with which he paired the socks in the drawer was quite admirable',
        'carol drank the blood as if she were a vampire',
        'giving directions that the mountains are to the west only works when you can see them',
        'the sign said there was road work ahead so he decided to speed up',
        'the gruff old man sat in the back of the bait shop grumbling to himself as he scooped out a handful of worms']
```

```python
countvec = CountVectorizer()
```

```python
countvec_fit = countvec.fit_transform(data)
```

```python
bag_of_words = pd.DataFrame(countvec_fit.toarray(), columns=countvec.get_feature_names_out())
```

```python
print(bag_of_words)
```

<h1 style="color:green">TF-IDF: Term Frequency Inverse Document Frequency</h1>

- #### TF-IDF keeps more of the context behind each word compared to Bag of Words
- #### measures how important a word is within a collection of texts
- #### Concern how often a word appears in a single document and how common it is across all documents
- ### Common words: lower score, Less Common words: higher score

```python
import pandas as pd 
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
tfidfvec = TfidfVectorizer()
```

```python
tfidfvec_fit = tfidfvec.fit_transform(data)
```

```python
tfidf_bag = pd.DataFrame(tfidfvec_fit.toarray(), columns = tfidfvec.get_feature_names_out())
```

```python
print(tfidf_bag)
```
