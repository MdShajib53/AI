## re.sub()
- searches for a specific text within a string and replaces it with new content
- re.sub("old text", "new text", "the string")

```python
string = r"sara was able to help me find the items I needed quickly"
```

```python
new_string = re.sub("sara", "sarah", string) 
```

```python
print(new_string)
```

```python
customer_reviews = ["sam was a great help to me in the store", 
                    "the cashier was very rude to me, I think her name was eleanor", 
                    "amazing work form sadeen!", 
                    "sarah was able to help me find the items I needed quickly",
                    "lucy is such a great addition to the team", 
                    "great service from sara she found me what i wanted"] 
```

```python
sarahs_reviews = []
```

```python
pattern_to_find = r"sarah?" # here ? means, 'h' is optional
```

```python
for string in customer_reviews: 
    if re.search(pattern_to_find, string): 
        sarahs_reviews.append(string)
```

```python
print(sarahs_reviews)
```

```python
a_reviews = []
pattern_to_find = r"^a" # r"^a" matches any string that begins with 'a' 

for string in customer_reviews: 
    if(re.search(pattern_to_find, string)): 
        a_reviews.append(string) 
```

```python
print(a_reviews)
```

```python
y_reviews = []
pattern_to_find = r"y$" # r"y$" mathces any string that end of the string 

for string in customer_reviews: 
    if(re.search(pattern_to_find, string)): 
        y_reviews.append(string) 
        
```

```python
print(y_reviews)
```

```python
needwant_reviews = []
pattern_to_find = r"(need|want)ed" # Match either needed or wanted
for string in customer_reviews:
    if re.search(pattern_to_find, string):
        needwant_reviews.append(string)
```

```python
print(needwant_reviews)
```

## Find the punctuation

```python
no_punct_reviews = [] 
pattern_to_find = r"[^\w\s]" # []: a set of characters we want to match, ^ means not, \w means word characters, \s means whitespaces 
# So, it means find anything that is not a word character and not a space 
```

```python
for string in customer_reviews: 
    no_punct_string = re.sub(pattern_to_find, "", string) 
    no_punct_reviews.append(no_punct_string)
```

```python
print(no_punct_reviews)
```

<h1 style='color:green'>Stemming</h1>

### Stemming: reducing words to their base form
- #### Reduces complexity
- #### Keeps the dataset easier to manage
###### Ex: connected-->connect

```python
from nltk.stem import PorterStemmer
```

```python
ps = PorterStemmer()
```

```python
connect_tokens = ['connecting', 'connected', 'connectivity', 'connect', 'connects']
```

```python
for t in connect_tokens: 
    print(t, ": ", ps.stem(t))
```

```python
learn_tokens = ['learned', 'learning', 'learn', 'learns', 'learner', 'learners']
```

```python
for t in learn_tokens: 
    print(t, ": ", ps.stem(t))
```

```python
likes_tokens = ['likes', 'better', 'worse']
```

```python
for t in likes_tokens: 
    print(t, ": ", ps.stem(t))
```

<h1 style='color:green'>Lemmatization</h1>

- ### Lemmatization: reduces a word to a meaningful base form while preserving its meaning 
- ### Lemmatization: uses a predefined dictionary to keep the context of the word

```python
import nltk
nltk.download('wordnet')
```

```python
from nltk.stem import WordNetLemmatizer
```

```python
lemmatizer = WordNetLemmatizer()
```

```python
for t in connect_tokens: 
    print(t,": ", lemmatizer.lemmatize(t))
```

```python
for t in learn_tokens: 
    print(t,": ", lemmatizer.lemmatize(t))
```

```python
for t in likes_tokens: 
    print(t,": ", lemmatizer.lemmatize(t))
```

<h1 style='color:green'>N-grams</h1>

- ### N-grams help us analyze the relationship between neighboring words
- ### N-gram = a sequence of N tokens
- ### N=1: unigrams ------> 'I', 'love', 'NLP'
- ### N=2: bigrams  ------> 'I love', 'love NLP'
- ### N=3: trigrams ------> 'I love NLP'

```python
import nltk 
import pandas as pd 
import matplotlib.pyplot as plt 
```

```python
tokens = ['the', 'rise', 'of', 'artificial', 'intelligence', 'has', 'led', 'to', 'significant', 'advancements', 'in', 'natural', 'language', 'processing', 'computer', 'vision', 'and', 'other', 'fields', 'machine', 'learning', 'algorithms', 'are', 'becoming', 'more', 'sophisticated', 'enabling', 'computers', 'to', 'perform', 'complex', 'tasks', 'that', 'were', 'once', 'thought', 'to', 'be', 'the', 'exclusive', 'domain', 'of', 'humans', 'with', 'the', 'advent', 'of', 'deep', 'learning', 'neural', 'networks', 'have', 'become', 'even', 'more', 'powerful', 'capable', 'of', 'processing', 'vast', 'amounts', 'of', 'data', 'and', 'learning', 'from', 'it', 'in', 'ways', 'that', 'were', 'not', 'possible', 'before', 'as', 'a', 'result', 'ai', 'is', 'increasingly', 'being', 'used', 'in', 'a', 'wide', 'range', 'of', 'industries', 'from', 'healthcare', 'to', 'finance', 'to', 'transportation', 'and', 'its', 'impact', 'is', 'only', 'set', 'to', 'grow', 'in', 'the', 'years', 'to', 'come']
print(tokens)
```

```python
unigrams = (pd.Series(nltk.ngrams(tokens, 1)).value_counts())
print(unigrams)
```

```python
unigrams[:10].sort_values().plot.barh(color='lightsalmon', width=0.9, figsize=(12,8))
plt.title("10 most frequently occuring Unigrams")
```

```python
bigrams = (pd.Series(nltk.ngrams(tokens, 2)).value_counts())
print(bigrams)
```

```python
bigrams[:10].sort_values().plot.barh(color='lightsalmon', width=0.9, figsize=(12,8))
plt.title("10 most frequently occuring Unigrams")
```

```python
trigrams = (pd.Series(nltk.ngrams(tokens, 3)).value_counts())
print(trigrams)
```

<h1 style="color:green">Practical Task - Text Preprocessing </h1>

```python
import nltk 
from nltk.tokenize import word_tokenize 
from nltk.stem import PorterStemmer, WordNetLemmatizer 
from nltk.corpus import stopwords
import re
import pandas as pd 
```

```python
data = pd.read_csv("tripadvisor-hotel-reviews.csv")
```

```python
data.info()
```

```python
data.head()
```

```python
data['Review'][0]
```

```python
data['review_lowercase'] = data['Review'].str.lower()
```

```python
data.head()
```

```python
en_stopwords = stopwords.words('english')
```

```python
en_stopwords.remove("not")
```

```python
data['review_no_stopwords'] = data['review_lowercase'] = data['review_lowercase'].apply(lambda x: ' '.join([word for word in x.split() if word not in (en_stopwords)]))
```

```python
data['review_no_stopwords'][0]
```

```python
data['review_no_stopwords_no_punct'] = data.apply(lambda x: re.sub(r"[*]", "star", x['review_no_stopwords']), axis=1)
```

```python
data.head()
```

```python
data['review_no_stopwords_no_punct'] = data.apply(lambda x: re.sub(r"([^\w\s])", "", x['review_no_stopwords_no_punct']), axis=1)
```

```python
data.head()
```

```python
data['tokenized'] = data.apply(lambda x: word_tokenize(x['review_no_stopwords_no_punct']), axis=1)
```

```python
data['tokenized'][0]
```

```python
ps = PorterStemmer()
```

```python
data['stemmed'] = data['tokenized'].apply(lambda tokens: [ps.stem(token) for token in tokens])
```

```python
data.head()
```

```python
lemmatizer = WordNetLemmatizer()
```

```python
data['lemmatized'] = data['tokenized'].apply(lambda tokens: [lemmatizer.lemmatize(token) for token in tokens])
```

```python
data['lemmatized'][0]
```

```python
tokens_clean = sum(data['lemmatized'], [])
```

```python
unigrams = (pd.Series(nltk.ngrams(tokens_clean, 1)).value_counts())
print(unigrams)
```

```python
bigrams = (pd.Series(nltk.ngrams(tokens_clean, 2)).value_counts())
print(bigrams)
```
