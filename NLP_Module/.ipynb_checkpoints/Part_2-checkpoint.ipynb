<h1 style="color:green">Lowercasing Text</h1>

```python
sentence = "Her cat's name is Luna"
```

```python
print(sentence)
```

```python
lower_sentence = sentence.lower()
```

```python
print(lower_sentence)
```

```python
sentence_list = ['Could you pass me the TV remote?',
                 'It is IMPOSSIBLE to find this hotel', 
                 'want to go for dinner on Tuesday?'] 
```

```python
lower_sentence_list = [x.lower() for x in sentence_list] 
print(lower_sentence_list)
```

<h1 style="color:green">Removing Stop Words</h1>

## Stopwords: and, of, a, to, ...

```python
import nltk
nltk.download('stopwords')
from nltk.corpus import stopwords
```

```python
en_stopwords = stopwords.words('english')
```

```python
print(en_stopwords)
```

```python
sentence = "it was far to go to the shop and he did not want her to walk"
```

```python
sentence_no_stopwords = ' '.join([word for word in sentence.split() if word not in en_stopwords])
```

```python
print(sentence_no_stopwords)
```

## re.sub()
- searches for a specific text within a string and replaces it with new content
- re.sub("old text", "new text", "the string")

```python
string = r"sara was able to help me find the items I needed quickly"
```

```python
new_string = re.sub("sara", "sarah", string) 
```

```python
print(new_string)
```

```python
customer_reviews = ["sam was a great help to me in the store", 
                    "the cashier was very rude to me, I think her name was eleanor", 
                    "amazing work form sadeen!", 
                    "sarah was able to help me find the items I needed quickly",
                    "lucy is such a great addition to the team", 
                    "great service from sara she found me what i wanted"] 
```

```python
sarahs_reviews = []
```

```python
pattern_to_find = r"sarah?" # here ? means, 'h' is optional
```

```python
for string in customer_reviews: 
    if re.search(pattern_to_find, string): 
        sarahs_reviews.append(string)
```

```python
print(sarahs_reviews)
```

```python
a_reviews = []
pattern_to_find = r"^a" # r"^a" matches any string that begins with 'a' 

for string in customer_reviews: 
    if(re.search(pattern_to_find, string)): 
        a_reviews.append(string) 
```

```python
print(a_reviews)
```

```python
y_reviews = []
pattern_to_find = r"y$" # r"y$" mathces any string that end of the string 

for string in customer_reviews: 
    if(re.search(pattern_to_find, string)): 
        y_reviews.append(string) 
        
```

```python
print(y_reviews)
```

```python
needwant_reviews = []
pattern_to_find = r"(need|want)ed" # Match either needed or wanted
for string in customer_reviews:
    if re.search(pattern_to_find, string):
        needwant_reviews.append(string)
```

```python
print(needwant_reviews)
```

## Find the punctuation

```python
no_punct_reviews = [] 
pattern_to_find = r"[^\w\s]" # []: a set of characters we want to match, ^ means not, \w means word characters, \s means whitespaces 
# So, it means find anything that is not a word character and not a space 
```

```python
for string in customer_reviews: 
    no_punct_string = re.sub(pattern_to_find, "", string) 
    no_punct_reviews.append(no_punct_string)
```

```python
print(no_punct_reviews)
```

<h1 style='color:green'>Tokenization</h1>

### Tokenization: breaking text into smaller units(tokens)

```python
import nltk 
nltk.download('punkt_tab') 
from nltk.tokenize import word_tokenize, sent_tokenize
```

```python
sentences = "Her cat's name is Luna. Her dog's name is Max"
```

```python
sent_tokenize(sentences)
```

```python
sentence = "her cat's name is luna"
word_tokenize(sentence)
```

```python
sentence2 = "Her cat's name is Luna and her dog's name is max"
word_tokenize(sentence2)
```
